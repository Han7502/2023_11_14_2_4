```python
#下載JSON的文字檔,json==>複雜的表格資料結構
import requests 
url = 'https://opendata.wra.gov.tw/Service/OpenData.aspx?format=json&id=F77C5EE5-2A30-4C62-B241-2F4CC8395864'

response = requests.get(url,verify=False) #verify=驗証,=False,全部不驗証
if response.ok:  #response.ok==>回應下載狀況
    print("下載成功")
    json_str:str = response.text
else:
    print("下載失敗")
```

```python
#解析json成為python的資料結構
import json
jsonData = json.loads(json_str)
jsonData
```

```python
#由response實體方法json(),直接轉換為python的資料結構
jsonData1 = response.json()
jsonData1['WaterResourcesAgencyRainfallStationsProfile_OPENDATA']
jsonRecords = jsonData1['WaterResourcesAgencyRainfallStationsProfile_OPENDATA']
```

```python
water_list:list[dict] = []
for item in jsonRecords:
    address_item:dict = {}
    address_item['位址'] = item['LocationAddress']
    address_item['橋名'] = item['ObservatoryName']
    water_list.append(address_item)
water_list
```

```python
import pandas as pd
dataFrame = pd.DataFrame(water_list)
dataFrame
```

```python
#dataFrame 可以使什麼?
#儲存為CSV
#儲存為excel

dataFrame.to_csv('淹水觀測點.csv')
```
