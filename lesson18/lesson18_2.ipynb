```python
import pandas as pd
#pd讀excel的指定columns
dataframe = pd.read_excel('year2021.xlsx',usecols=['國家','日期','總確診數','新增確診數'])
dataframe
```

```python
#利用query搜尋
dataframe.query('新增確診數>1000') 
```

```python
#針對特定column排序, sort_values(ascending=True==>由小到大,by:str)
#有了排序可以找最大或最小的前幾筆
sortedFrame = dataframe.sort_values(by='新增確診數',ascending=False)
#.iloc[:?]找出前幾筆
sortedFrame.iloc[:20]
```
