```python
import pandas as pd
#將csv直接轉換成dataframe
DataFrame1 = pd.read_csv('各鄉鎮市區人口密度.csv')
DataFrame1
```

```python
#取出columns,index
DataFrame1.index
DataFrame1.columns
```

```python
#改變欄位名稱
DataFrame1.columns = ['統計年','區域別','年底人口數','土地面積','人口密度']

DataFrame1
```

```python
#使用pandas.dataframe.drop()直接刪除列或reindex()適合選取特定列,刪除資料
dataframe2 = DataFrame1.drop(index=0)
dataframe2
```

```python
DataFrame1.reindex(index=[1,2,3,4,5])
```

```python
dataframe2.iloc[-10:]#查看倒數5行
dataframe3 = dataframe2.reindex(index=range(1,371))
#用reindex(index=range(?,?)存取我們要的列數
```

```python
dataframe3
```

```python
#選取留存所要的欄位
dataframe4 = dataframe3.reindex(columns=['區域別','年底人口數','土地面積','人口密度'])
dataframe4
```

```python
dataframe5 = dataframe4.drop(columns=['區域別'])
dataframe5
```
